# EDA - Medicine Recommendation Dataset

```python
import sys
sys.path.append("../src")
import pandas as pd
import plotly.express as px

from preprocessing import load_raw_data, clean_data

df = clean_data(load_raw_data("../data/Cleaned_Dataset.csv"))
df.head()
```

## Top diseases

```python
top = df["disease"].value_counts().head(15)
px.bar(top, orientation="h")
```

## Patient distribution (age & gender)

```python
px.histogram(df, x="age", color="gender", nbins=20)
```

## Risk level popularity

```python
risk_counts = df["risk_level"].value_counts()
px.pie(values=risk_counts.values, names=risk_counts.index)
```

## Symptom frequency

```python
symptom_cols = ["fever_flag", "cough_flag", "fatigue_flag", "difficulty_breathing_flag"]
symptom_freq = df[symptom_cols].sum()
px.bar(symptom_freq)
```

## Correlation heatmap

```python
corr = df[["age", "blood_pressure", "cholesterol_level"]].corr()
px.imshow(corr, text_auto=True, color_continuous_scale="Blues")
```
